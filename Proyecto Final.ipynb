```python
net_manager: Código del gestor de la red regional.
purchase_area: Código del área donde se compra la energía.
street: Nombre de la calle.
zipcode_from y zipcode_to: Dos columnas que indican el rango de códigos postales cubiertos, compuesto por 4 números y 2 letras.
city: Nombre de la ciudad.
num_connections: Número de conexiones en el rango de códigos postales.
delivery_perc: Porcentaje del consumo neto de electricidad o gas. Cuanto más bajo sea este porcentaje, más energía se ha devuelto a la red (por ejemplo, si tienes paneles solares).
perc_of_active_connections: Porcentaje de conexiones activas en el rango de códigos postales.
type_of_connection: Tipo principal de conexión en el rango de códigos postales. Para electricidad es el número de fusibles y amperios. Para gas son las categorías G4, G6, G10, G16, G25.
type_conn_perc: Porcentaje de presencia del tipo principal de conexión en el rango de códigos postales.
annual_consume: Consumo anual. En kWh para electricidad y en m³ para gas.
annual_consume_lowtarif_perc: Porcentaje de consumo durante las horas de tarifa baja. De 10 p.m. a 7 a.m. y durante los fines de semana.
smartmeter_perc: Porcentaje de medidores inteligentes en los rangos de códigos postales. Aparato en la casa que calcula precisamente cuánto se usa de electricidad en porcentaje por metro en los rangos de códigos postales.
```

```python
import pandas as pd
import folium
from folium.plugins import HeatMap
import requests
import webbrowser

# Reemplaza 'YOUR_OPENCAGE_API_KEY' con tu clave de API válida
api_key = 'd7c158ac69df4f0f9963430e08a7ea83'

# Diccionario para almacenar las coordenadas en caché
coordinates_cache = {}

# Función para obtener coordenadas de un código postal usando la API de OpenCage
def get_coordinates(postal_code):
    if postal_code in coordinates_cache:
        return coordinates_cache[postal_code]
    
    url = f'https://api.opencagedata.com/geocode/v1/json?q={postal_code}&key={api_key}'
    response = requests.get(url)
    if response.status_code == 200:
        results = response.json()['results']
        if results:
            coordinates = (results[0]['geometry']['lat'], results[0]['geometry']['lng'])
            coordinates_cache[postal_code] = coordinates
            return coordinates
    return None, None

# Supongamos que df_select2 es tu DataFrame con los datos necesarios
# Crear una nueva columna 'zipcode_range' combinando 'zipcode_from' y 'zipcode_to'
df_select2['zipcode_range'] = df_select2['zipcode_from'] + '-' + df_select2['zipcode_to']

# Agrupar por código postal único y obtener coordenadas para cada grupo
unique_zipcodes = df_select2['zipcode_from'].unique()
coords_dict = {zipcode: get_coordinates(zipcode) for zipcode in unique_zipcodes}

# Aplicar las coordenadas obtenidas a todo el conjunto de datos
df_select2['latitude'] = df_select2['zipcode_from'].map(lambda x: coords_dict[x][0])
df_select2['longitude'] = df_select2['zipcode_from'].map(lambda x: coords_dict[x][1])

# Filtrar filas con coordenadas válidas para toda la población
df_select2 = df_select2.dropna(subset=['latitude', 'longitude'])

# Verificar los datos procesados
print("Datos de toda la población:")
print(df_select2)

# Verificar que hay filas válidas después de filtrar
if not df_select2.empty:
    # Crear un mapa base centrado en una ubicación media para toda la población
    map_center_all = [df_select2['latitude'].mean(), df_select2['longitude'].mean()]
    m_all = folium.Map(location=map_center_all, zoom_start=10, tiles='CartoDB positron')

    # Añadir un mapa de calor para toda la población basado en rangos de zonas
    heat_data_all = [[row['latitude'], row['longitude'], row['total_annual_consume']] for index, row in df_select2.iterrows()]
    HeatMap(heat_data_all, radius=15, blur=10, max_zoom=1).add_to(m_all)

    # Guardar el mapa en un archivo HTML para toda la población
    m_all.save('heatmap_zipcodes_all.html')
    webbrowser.open('heatmap_zipcodes_all.html')
else:
    print("No hay datos válidos para toda la población.")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import xgboost as xgb
import lightgbm as lgb
from sklearn.ensemble import GradientBoostingRegressor
import numpy as np  # Asegúrate de importar numpy

# Supongamos que df_select2 es tu DataFrame
Q1 = df_select2['total_annual_consume'].quantile(0.25)
Q3 = df_select2['total_annual_consume'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filtrar los outliers
df_filtered = df_select2[(df_select2['total_annual_consume'] >= lower_bound) & 
                         (df_select2['total_annual_consume'] <= upper_bound)]

X_filtered = df_filtered[['num_connections', 'delivery_perc', 'perc_of_active_connections', 'type_conn_perc', 'latitude','longitude']]
y_filtered = df_filtered['total_annual_consume']

# Dividir en conjuntos de entrenamiento y prueba
X_train_filtered, X_test_filtered, y_train_filtered, y_test_filtered = train_test_split(X_filtered, y_filtered, test_size=0.2, random_state=42)

# Modelos
models = {
    "XGBoost": xgb.XGBRegressor(n_estimators=100, learning_rate=0.1, random_state=42),
    "LightGBM": lgb.LGBMRegressor(n_estimators=100, learning_rate=0.1, random_state=42),
    "Gradient Boosting": GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, random_state=42)
}

# Entrenamiento y evaluación
results = {}
for name, model in models.items():
    model.fit(X_train_filtered, y_train_filtered)
    y_pred = model.predict(X_test_filtered)
    mse = mean_squared_error(y_test_filtered, y_pred)
    rmse = np.sqrt(mse)  # Calcular el RMSE
    r2 = r2_score(y_test_filtered, y_pred)
    results[name] = {'RMSE': rmse, 'R^2': r2}  # Cambiar MSE por RMSE

# Mostrar resultados
for name, metrics in results.items():
    print(f"{name} - RMSE: {metrics['RMSE']}, R^2: {metrics['R^2']}")

# Visualización de la distribución filtrada
sns.histplot(df_filtered['total_annual_consume'], bins=30, kde=True)
plt.title('Distribución de Total Annual Consume (Filtrada)')
plt.xlabel('Total Annual Consume')
plt.ylabel('Frecuencia')
plt.show()
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import mean_squared_error, r2_score
import lightgbm as lgb

class HyperparameterTuner:
    def __init__(self, estimator, param_distributions, n_iter, cv, n_jobs, verbose, random_state):
        self.estimator = estimator
        self.param_distributions = param_distributions
        self.n_iter = n_iter
        self.cv = cv
        self.n_jobs = n_jobs
        self.verbose = verbose
        self.random_state = random_state

    def tune_hyperparameters(self, X, y):
        self.random_search = RandomizedSearchCV(
            estimator=self.estimator,
            param_distributions=self.param_distributions,
            n_iter=self.n_iter,
            cv=self.cv,
            n_jobs=self.n_jobs,
            verbose=self.verbose,
            random_state=self.random_state
        )
        self.random_search.fit(X, y)
        self.best_estimator = self.random_search.best_estimator_
        self.best_params = self.random_search.best_params_
        return self.best_estimator, self.best_params

    def evaluate_model(self, X_test, y_test):
        y_pred = self.best_estimator.predict(X_test)
        mse = mean_squared_error(y_test, y_pred)
        r2 = r2_score(y_test, y_pred)
        return mse, r2

# Definir los hiperparámetros a buscar
param_dist_lgb = {
    'num_leaves': [31, 50, 70, 100],
    'learning_rate': [0.01, 0.05, 0.1, 0.2],
    'n_estimators': [100, 200, 300, 400, 500],
    'max_depth': [-1, 5, 10, 15, 20],
    'min_child_samples': [20, 30, 40, 50],
    'subsample': [0.6, 0.7, 0.8, 0.9, 1.0],
    'colsample_bytree': [0.6, 0.7, 0.8, 0.9, 1.0],
    'reg_alpha': [0, 0.01, 0.1, 1],
    'reg_lambda': [0.5, 1, 1.5, 2]
}

# Crear el modelo LightGBM
lgb_model = lgb.LGBMRegressor(random_state=42)

# Crear y entrenar el modelo
tuner = HyperparameterTuner(lgb_model, param_dist_lgb, 100, 3, -1, 2, 42)
best_model, best_params = tuner.tune_hyperparameters(X_train_filtered, y_train_filtered)

# Imprimir los mejores parámetros
print("Best Parameters found by RandomizedSearchCV:")
print(best_params)

# Evaluar el modelo
mse, r2 = tuner.evaluate_model(X_test_filtered, y_test_filtered)

# Calcular RMSE
rmse = mse ** 0.5  # Calculando RMSE a partir de MSE

# Imprimir las métricas de evaluación
print(f'Random Search Best LightGBM RMSE: {rmse:.2f}')
print(f'Random Search Best LightGBM R^2: {r2:.2f}')

```
